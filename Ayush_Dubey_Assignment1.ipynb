```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = int(input("Enter distance r :"))
d=int(input("Enter no. of  dimensions d :"))
nd = d 
nop = 1000
p = np.random.rand(nop, nd)*100
print(p)
```

```python
def adj(p1, p2):
    for i in range(len(p1)):
        if abs(p1[i] - p2[i]) > 1:
            return False
    return True
```

```python
def dist(p1, p2):
    return np.sqrt(np.sum((p1 - p2)**2))
```

```python
def gridco(p):
    return np.floor(p).astype(int)
```

```python
k = set()


for i in range(0, len(p)):
    a = p[i]
    for j in range(i + 1, len(p)):
        b = p[j]
        if adj(a, b):
            d = dist(a, b)
            if d < r:
                k.add((tuple(a), tuple(b)))
        else:
            continue
```

```python
print("Set A = ")
print(k)
```

```python
poi=np.array(list(k))
plt.scatter(poi[:,0],poi[:,1],c='blue',marker='o')
plt.show()
```
